## Gravity field plots

```julia
include("gravityfield.jl")
```

Define a 3D planet system, plot gravity magnitude parallel to the planet plane 

```julia
earth = Sphere()
moon = earthlikeSphere_m_known(p = [d_moon_earth, 0., 0.], m = m_moon)
sun = Sphere([-d_sun_earth, 0., 0.], m_sun, (3m_sun / (4 * _sun))^(1/3))
bodies = [sun, earth, moon]

xgrid, ygrid = xygrids( xmin = -d_moon_earth/2., 
                        xmax = d_moon_earth * 3/2, 
                        ymin =-d_moon_earth/2, 
                        ymax = d_moon_earth * 3/2, n = 100)

gmagn = gfieldvals(bodies, xgrid, ygrid, zplane = moon.r)
plotfield(xgrid, ygrid, gmagn)
```

```julia
pygui(true)
plotfield(xgrid, ygrid, gmagn)
```
