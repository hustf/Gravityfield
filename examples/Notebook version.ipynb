# Test plotting

Let's first test that plotting works. We're using PyPlot because it supports streamplot with colors, whereas the versatile Plots interface doesn't.

```julia
let
    n = 100
    u = range(0, 2π, length = n)
    v = range(0, π, length = n)
    x = cos.(u) * sin.(v)'m
    y = sin.(u) * sin.(v)'m
    z = ones(n) * cos.(v)'m
    np = nest(x, y, z)
    wrap_surface_and_countour(np, title = "Just a sphere")
end
```
